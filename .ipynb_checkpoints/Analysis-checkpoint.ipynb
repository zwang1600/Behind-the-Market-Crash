# Relationships Between Sector Trends And the General Trend of S&P 500 Index During the 2020 Stock Market Crush

### Determine The Time Period

In this notebook, we will investigate the relationships between individual stock prices and the general [S&P 500 index](https://en.wikipedia.org/wiki/S%26P_500) during the [2020 stock market crash](https://en.wikipedia.org/wiki/2020_stock_market_crash).

Thanks to Wikipedia, we can easily find out the date range of the crash - from 02/20/2020 to 04/07/2020. However, since we are interested in the impact of the crash, we also need to pay attention to the dates prior and after the crash. Therefore, we should set the the time period of the investigation to ``02/10/2020 - 04/17/2020``(10 days prior and after the crash.)

### Gathering Index Data

The [dataset](https://www.kaggle.com/arashnic/time-series-forecasting-with-yahoo-stock-price) we are using for the index is provided by [Kaggle](https://www.kaggle.com/) and Yahoo. Kaggle is an online data science community and we can download the dataset from it with little trouble. However, the yahoo_stock.csv contains the data from 2015-2020, so we need to do a little bit of trimming before going forward:

```python
import csv
from datetime import datetime

start = "2020-02-10"
end = "2020-04-17"

with open('Data/yahoo_stock.csv') as f:
    with open('Data/yahoo_stock_trimed.csv', "w") as r: 
        reader = csv.reader(f)  
        writer = csv.writer(r)
        writer.writerow(["Date", "High", "Low", "Open", "Close", "Volume", "Adj Close"])
        for row in reader:
            if (row[0] >= start and row[0] <= end):
                writer.writerow(row)
```

### Visualize the General Trend
Now we have the data during the desired time period. And let's see what the graph looks like:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Data/yahoo_stock_trimed.csv', parse_dates = ['Date'], index_col = 'Date')
df.head()

plt.figure(figsize=(15,5))
plt.plot(df.index, df.Open, lw=2, color='b',label='Open')
plt.plot(df.index, df.Close, lw=2, color='r',label='Close')
plt.plot(df.index, df.High, lw=1, color='g',linestyle='--',alpha=0.7,label='High')
plt.plot(df.index, df.Low, lw=1, color='m',linestyle='--',alpha=0.7,label='Low')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.show()
```

Above is the visualization of all the metrics of the S&P 500 index:

   **High** -> Highest Price of the stock for that particular date.

   **Low** -> Lowest Price of the stock for that particular date.

   **Open** -> Opening Price of the stock.

   **Close** -> Closing Price of the stock.
   
Since our focus on the general trend of the index, instead of the small difference between each measurement, let's just look at an individual metric (close price) for the upcoming analysis.

```python
plt.figure(figsize=(15,5))
plt.plot(df.index, df.Close, lw=2, color='r',label='Close')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.show()
```

### Gathering Individual Stock Data

Once again, the [dataset](https://www.kaggle.com/nroll12/sp-500-full-dataset) we are using for individual stock price also contains more data point than we need(1999-2020.) So we have to trim the datasets as well.

```python
import os
import pandas as pd

for file in os.listdir('Data/Raw-Individual-Stock-Data'):
    df = pd.read_csv('Data/Raw-Individual-Stock-Data/' + file)
    # Select the desired dates
    df = df.loc[(df['Date'] >= start) & (df['Date'] <= end)]
    # Drop the "Dividends" and "Stock Splits" column since we don't need them in this notebook
    df = df.drop("Dividends",1)
    df = df.drop("Stock Splits",1)
    # Store the output in the "Trimed-Individual-Stock-Data" folder
    df.to_csv('Data/Trimed-Individual-Stock-Data/' + file, index = False)
# Let us know that this process is finished
print("Done")
```

After triming the individual stock data, let's visualize a few csv files just to make sure we have the correct data!

```python
import random

# Graph three random files
for x in range(3):
    f = random.choice(os.listdir('Data/Trimed-Individual-Stock-Data'))
    df = pd.read_csv('Data/Trimed-Individual-Stock-Data/'+f, parse_dates = ['Date'], index_col = 'Date')
    plt.figure(figsize=(15,5))
    plt.plot(df.index, df.Close, lw=2, color='r',label='Close')
    plt.title('Stock Symbol: '+f[:len(f) - 4])
    plt.legend()
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.show()
```

### Categorizing the Stocks
Each different stock belongs to different sector (e.x. Health Care, Information Technology, Utilies, etc.) [This dataset](https://www.kaggle.com/paytonfisher/sp-500-companies-with-financial-information) offers us financial insights of each individual stocks that belong to the S&P 500 index. So let's use the information provided to categorize the stock data we just obtained!

```python
# pretty print for our list
from pprint import pprint

# the list to store all the different sector names
sectors = []
# the dataframe from the financial.csv file
financials = pd.read_csv('Data/financials.csv')

# iterate all the rows in the dataframe
for index, row in financials.iterrows():
    # if our list doesn't contain this sector, add it to the list
    if row[2] not in sectors:
        sectors.append(row[2])
# print the list after iterating over the whole dataframe
pprint(sectors)
print(len(sectors))
```

So we have 11 different sectors: Industrials, Health Care, Information Technology, Consumer Discretionary, Utilities, Financials, Materials, Real Estate, Consumer Staples, Energy, and Telecommunication Services.

We will then move each individual stock into the sector folder they belong. First, we need to create all the sub-folders needed:

```python
for sector in sectors:
    path = 'Data/Trimed-Individual-Stock-Data/'+sector
    # make new directory if the directory doesn't exsist already
    if not os.path.isdir(path):
        os.makedirs(path)
```

After creating all the sub-folders, we then move all the csv files according to their sector.

```python
import shutil

for index, row in financials.iterrows():
    filename = row[0].lower() + '.csv'
    source = 'Data/Trimed-Individual-Stock-Data/'
    destination = 'Data/Trimed-Individual-Stock-Data/' + row[2] + '/'
    # if the symbol exsist in our folder then we move it to the subfolder it belongs to
    if os.path.isfile(source + filename):
        shutil.move(source + filename, destination + filename)
print('Done')
```

As we can see, we still have some "left over" stocks that don't belong to any of the sector. And that's because the two datasets we are using contain slight different symbols. But as long as the majority of the stocks are categorized, we should be able to move forward. Let's see how many stocks we have in each sectors:

```python
notcategorized = 0
# iterate over all the files in the folder
for f in os.listdir('Data/Trimed-Individual-Stock-Data/'):
    # this is a file, which means this stock is not categorized 
    if os.path.isfile(source + f):
        notcategorized += 1
    # this is a folder
    else:
        count = 0;
        # count the number of files in the subfolder
        for f2 in os.listdir('Data/Trimed-Individual-Stock-Data/' + f + '/'):
            count += 1
        print(f + ': ' + str(count))

print('Not categorized: ' + str(notcategorized))
```

### Analyze Stocks Based on Sector

Now that we have everything categorized, we can see if any sectors responded differently to the market crash. We can accomplish this by graphing the general index trend along with all the stock trends within the sector. However, since every stock has an unique price range and the difference can be really dramatic, we need to overlook the price and only investigate the trend. Fortunally, [matplotlib twinx](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.twinx.html) can provide the tools needed for this task.

```python
# root folder
folder = 'Data/Trimed-Individual-Stock-Data/'

for subfolder in os.listdir(folder):
    # this is a sector folder
    if os.path.isdir(folder + subfolder):
        # the path of the sector folder
        path = folder + subfolder + '/'
        fig, host = plt.subplots(figsize=(15,5))
        plt.xlabel('Date')
        plt.title(subfolder)

        count = 0
        # loop over all the files in the sector folder
        for f in os.listdir(path):
            count += 1
            df = pd.read_csv(path + f, parse_dates = ['Date'], index_col = 'Date')
            # create a 'twin' of the host so that the difference in stock price will be overlooked
            par = host.twinx()
            # plot the 'twin'
            par.plot(df.index, df.Close, lw=2, color='r',label=f[:len(f) - 4])
            # make the y-axis invisible since it makes the graph look messy
            par.axes.get_yaxis().set_visible(False)
        
        # plot the general index trend in black
        df = pd.read_csv('Data/yahoo_stock_trimed.csv', parse_dates = ['Date'], index_col = 'Date')
        host.plot(df.index, df.Close, lw=2, color='k',label=f[:len(f) - 4])
        host.axes.get_yaxis().set_visible(False)
        plt.show()
```

### Analysis

As we can see from the visualization above, while the general trend of most of the stocks appear to be downward and somewhat follow the general index trend, there are difference in between different sectors. 

The first takeaway is that stocks in each sector appear to have silimar trends. For instance, in the "utilities" or the "fiancials" sector, all the trendlines have very simiar shapes. However, there are outliers to this obeservation. For instance, there is one stock in the "energy" sector that started low in late Febuary and early March and then had a little spkied in mid-March while the rest of the energy stocks were plummeting. And in the "Consumer Discretionary" sector, there is one stock that didn't follow the pattern of the rest of the stocks in that category in between 03/15 and 04/15. But given the number of outliers in comparsion to the totoal number of stocks, it is safe to say that most stocks in the same sector do tend to follow the same pattern.

Another takeaway is that, suprisingly, different sectors seem to react to the market crush more or less the same. Specifically, all the sectors had a dip at the end of Feburary, then they gradually decline from the beginning of March to mid-March, and there were two small spikes for all of them on around March 12th and March 16th. Even though the degree of the spikes are slightly different -- for instance, the spike on March 16th is relatively large for utilities and cousmer staples, but smaller for the rest of the sectors -- the shape around that period is oddly similar among all categories. And of course, the price rose gradually from late March to mid-April in smilar fashions for all the sectors. 

And in order to see if we made the right observation, let's investigate more in details about the trends in between different sectors.

```python
# root folder
folder = 'Data/Trimed-Individual-Stock-Data/'

# plot the general index trend in black
fig, host = plt.subplots(figsize=(15,5))
plt.xlabel('Date')
df = pd.read_csv('Data/yahoo_stock_trimed.csv', parse_dates = ['Date'], index_col = 'Date')
# we create a different dataframe to store data for later usage
df2 = pd.DataFrame()
df2['Date'] = df.index
df2['Index'] = df['Close'].values
df2 = df2.drop([5,6,7,12,13,19,20,26,27,33,34,40,41,47,48,54,55,60,61,62])
host.plot(df.index, df.Close, lw=2, color='k',label='General Index')
host.axes.get_yaxis().set_visible(False)
host.legend(loc="lower left", bbox_to_anchor=(0,0))

colors = ['#ff3333', '#ff9c33', '#ffe933', '#a5ff33', '#33ffc1', '#33ceff', '#3361ff', '#be33ff', '#ff33d7', '#00ffae', '#ff0000']
count = 0
    
for subfolder in os.listdir(folder):
    # this is a sector folder
    if os.path.isdir(folder + subfolder):
        # the path of the sector folder
        path = folder + subfolder + '/'
        
        # Initialize a dictionary of size 48 because there are 48 days in the date range
        data = {}
        for x in range(48):
            data.update({x: 0})
        
        # traverse through all the files in the sector 
        for f in os.listdir(path):
            df = pd.read_csv(path + f, parse_dates = ['Date'])
            # traverse through the rows in each file
            for index, row in df.iterrows():
                # add the data to the corresponding date
                data[index] += row['Close']
        
        # store the data for later usage
        df2[subfolder] = data.values()
        # make a twin graph of the host
        par = host.twinx()
        # make the y-axis invisible since it makes the graph look messy
        par.axes.get_yaxis().set_visible(False)
        # graph the line
        par.plot(df['Date'], list(data.values()), lw=2, color=colors[count],label=subfolder)
        # offset the legend
        par.legend(loc="lower left", bbox_to_anchor=(0, count * 0.06 + 0.06))
        count += 1
    
plt.show()
df2.to_csv('Data/sector_sum_trend.csv')
```

Throughout the process, we stored the data in the csv file titled ``sector-sum-trend.csv``. That's dived more in depth and analyze this csv file!

```python
df = pd.read_csv('Data/sector_sum_trend.csv')
df2 = pd.DataFrame()

data = {'Index': [None] * 47,
        'Industrials': [None] * 47,
        'Real Estate': [None] * 47,
        'Energy': [None] * 47,
        'Information Technology': [None] * 47,
        'Utilities': [None] * 47,
        'Telecommunication Services': [None] * 47,
        'Financials': [None] * 47,
        'Consumer Discretionary': [None] * 47,
        'Materials': [None] * 47,
        'Consumer Staples': [None] * 47,
        'Health Care': [None] * 47,
       }

for index, row in df.iterrows():
    if index == 0:
        prev = row
    else:
        curr = row
        # Okay, I know this is redundant. But since there're only 12 of them, so...
        # And these are to calculate the fluctuation ((current day - previous day) / previous day)
        data['Index'][index - 1] = ("{:.3%}".format((curr['Index'] - prev['Index'])/prev['Index']))
        data['Industrials'][index - 1] = ("{:.3%}".format((curr['Industrials'] - prev['Industrials'])/prev['Industrials']))
        data['Real Estate'][index - 1] = ("{:.3%}".format((curr['Real Estate'] - prev['Real Estate'])/prev['Real Estate']))
        data['Energy'][index - 1] = ("{:.3%}".format((curr['Energy'] - prev['Energy'])/prev['Energy']))
        data['Information Technology'][index - 1] = ("{:.3%}".format((curr['Information Technology'] - prev['Information Technology'])/prev['Information Technology']))
        data['Utilities'][index - 1] = ("{:.3%}".format((curr['Utilities'] - prev['Utilities'])/prev['Utilities']))
        data['Telecommunication Services'][index - 1] = ("{:.3%}".format((curr['Telecommunication Services'] - prev['Telecommunication Services'])/prev['Index']))
        data['Financials'][index - 1] = ("{:.3%}".format((curr['Financials'] - prev['Financials'])/prev['Financials']))
        data['Consumer Discretionary'][index - 1] = ("{:.3%}".format((curr['Consumer Discretionary'] - prev['Consumer Discretionary'])/prev['Consumer Discretionary']))
        data['Materials'][index - 1] = ("{:.3%}".format((curr['Materials'] - prev['Materials'])/prev['Materials']))
        data['Consumer Staples'][index - 1] = ("{:.3%}".format((curr['Consumer Staples'] - prev['Consumer Staples'])/prev['Consumer Staples']))
        data['Health Care'][index - 1] = ("{:.3%}".format((curr['Health Care'] - prev['Health Care'])/prev['Health Care']))
        prev = curr

print(data)

# with open('Data/sector-fluctuation-data.csv', 'w') as f:
#     writer = csv.writer(f)
    
```
